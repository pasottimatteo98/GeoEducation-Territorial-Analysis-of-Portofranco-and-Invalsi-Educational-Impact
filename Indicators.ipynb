# Ben Banchi
- rosa = ambito personale
- verde = ambito sociale
- giallo = ambito di interesse (ben banchi --> solidarietà, portofranco --> scolastico)

```python
Ben_Banchi_iniziale = pd.read_csv("/content/gdrive/MyDrive/Data Science Lab/dataset/Beneficiari Banchi_final.csv")
var=[]
for i in Ben_Banchi_iniziale:
   var.append(i)
split=var[6:-4]
somma=var[5:-4]
Ben_Banchi=Ben_Banchi_iniziale[split]
```

```python
# ambito personale
pers=[]
pers.append(var[6])
pers.append(var[8])
pers.append(var[9])
pers.append(var[12])
pers.append(var[13])
pers.append(var[15])
pers.append(var[16])
pers.append(var[19])
ambito_personale = pd.DataFrame()
for i in pers:
  ambito_personale[i]=Ben_Banchi[i]
ambito_personale
```

```python
ambito_personale_mediana=[]
for i in ambito_personale.iterrows():
  ambito_personale_mediana.append(i[1].median())
```

```python
# ambito sociale
soc=[]
soc.append(var[7])
soc.append(var[20])
soc.append(var[21])
soc.append(var[23])
ambito_sociale = pd.DataFrame()
for i in soc:
  ambito_sociale[i]=Ben_Banchi[i]
ambito_sociale
```

```python
ambito_sociale_mediana=[]
for i in ambito_sociale.iterrows():
  ambito_sociale_mediana.append(i[1].median())
```

```python
# ambito solidarietà
sol=[]
sol.append(var[10])
sol.append(var[11])
sol.append(var[14])
sol.append(var[17])
sol.append(var[18])
sol.append(var[22])
ambito_solidarietà = pd.DataFrame()
for i in sol:
  ambito_solidarietà[i]=Ben_Banchi[i]
```

```python
ambito_solidarietà_mediana=[]
for i in ambito_solidarietà.iterrows():
  ambito_solidarietà_mediana.append(i[1].median())
```
